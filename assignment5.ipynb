```python
# 2) DATABASE Practice using SQLite

import sqlite3

# Create database
conn = sqlite3.connect("college.db")

# Create cursor
cur = conn.cursor()

# Create tables
cur.execute("CREATE TABLE IF NOT EXISTS Student(id INTEGER, name TEXT, marks INTEGER)")
cur.execute("CREATE TABLE IF NOT EXISTS Teacher(id INTEGER, name TEXT, subject TEXT)")
cur.execute("CREATE TABLE IF NOT EXISTS Course(id INTEGER, course_name TEXT)")

# Insert records
cur.execute("INSERT INTO Student VALUES(1, 'Rahul', 85)")
cur.execute("INSERT INTO Student VALUES(2, 'Anjali', 90)")

cur.execute("INSERT INTO Teacher VALUES(1, 'Sharma', 'Math')")
cur.execute("INSERT INTO Teacher VALUES(2, 'Verma', 'Science')")

cur.execute("INSERT INTO Course VALUES(1, 'Python')")
cur.execute("INSERT INTO Course VALUES(2, 'Database')")

conn.commit()

# Different Select Operations
print("Student Table:")
cur.execute("SELECT * FROM Student")

for row in cur.fetchall():
    print(row)

print("\nStudents with marks greater than 85:")
cur.execute("SELECT * FROM Student WHERE marks > 85")

for row in cur.fetchall():
    print(row)

# Update data
cur.execute("UPDATE Student SET marks = 95 WHERE name='Rahul'")
conn.commit()

print("\nAfter Update:")
cur.execute("SELECT * FROM Student")

for row in cur.fetchall():
    print(row)

# Delete data
cur.execute("DELETE FROM Student WHERE id=2")
conn.commit()

print("\nAfter Delete:")
cur.execute("SELECT * FROM Student")

for row in cur.fetchall():
    print(row)

# Close connection
conn.close()
print("\nDatabase operations completed.")
```

```python
# 1) Create a CSV file for address book

import csv

# CSV file name
file = "address_book.csv"

# Column names
fields = ["Name", "Address", "Mobile", "Email"]

# Empty list to store data
data = []

# Taking input from user
n = int(input("How many records you want to enter: "))

for i in range(n):
    print("\nEnter details of person", i + 1)

    name = input("Enter Name: ")
    address = input("Enter Address: ")
    mobile = input("Enter Mobile: ")
    email = input("Enter Email: ")

    data.append([name, address, mobile, email])

# Writing data into CSV file
with open(file, "w", newline="") as f:
    writer = csv.writer(f)

    writer.writerow(fields)   # Write column names
    writer.writerows(data)    # Write user data

print("\nData stored successfully in address_book.csv")
```
