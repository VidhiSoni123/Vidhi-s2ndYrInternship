```python
#assignment 1-5th question
name=input("enter name of student")
print(name)
clas=int(input("enter class of student"))
print(clas)
Total_Marks=0
for x in range(1,6,1):
    marks=int(input("enter marks out of 100"))
    Total_Marks=Total_Marks+marks
print(Total_Marks) 
p=(Total_Marks/500)*100
print(p)

if(100>p>=60):
    print("rankA")
elif(60>p>=50):
    print("RankB")
elif(50>p>=40):
    print("RankC")
elif(40>p>=33):
    print("RankD")
else:
    print("Fail")
```
