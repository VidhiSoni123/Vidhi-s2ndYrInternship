```python
# 9. Count upper and lower case letters

def count_case(text):
    upper = 0
    lower = 0

    for i in text:
        if i.isupper():
            upper += 1
        elif i.islower():
            lower += 1

    print("Upper case letters:", upper)
    print("Lower case letters:", lower)

count_case("PyThOn")
```

```python
# 8. Check prime number

def prime(n):
    if n <= 1:
        print("Not Prime")
    else:
        for i in range(2, n):
            if n % i == 0:
                print("Not Prime")
                break
        else:
            print("Prime")

prime(7)
```

```python
# 7. Print even numbers from list

def even_numbers(lst):
    for i in lst:
        if i % 2 == 0:
            print(i)

even_numbers([1, 2, 3, 4, 5, 6])
```

```python
# 6. Check number within range

def check_range(num):
    if num in range(1, 11):
        print("Number is in range")
    else:
        print("Number is not in range")

check_range(5)
```

```python
# 5. Reverse a string

def reverse_string(text):
    return text[::-1]

print(reverse_string("Python"))
```

```python
# 4. Factorial of a number

def factorial(n):
    fact = 1
    
    for i in range(1, n + 1):
        fact = fact * i
    
    return fact

print(factorial(5))
```

```python
# 3. Multiply all numbers in a list

def multiply_list(lst):
    result = 1
    
    for i in lst:
        result = result * i
    
    return result

print(multiply_list([2, 3, 4]))
```

```python
# 2. Distinct elements from list

def distinct_list(lst):
    new_list = []
    
    for i in lst:
        if i not in new_list:
            new_list.append(i)
    
    return new_list

print(distinct_list([1, 2, 2, 3, 4, 4, 5]))
```

```python
# 1. Maximum of three numbers

def maximum(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

print(maximum(10, 25, 15))
```
